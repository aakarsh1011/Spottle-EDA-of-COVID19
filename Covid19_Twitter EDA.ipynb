```python
#importing necessary libraries
import numpy as np
import pandas as pd
import warnings 
import re

#importing data as a text file
filename = 'D:/work/data science/tweets_corona.txt'
```

```python
#libraries for visualization
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS


#ignoring warnings int he output
warnings.filterwarnings('ignore')
```

```python
file = open(filename,encoding='utf8')
text = file.read()
tweets = text.split("\n")
```

```python
len(tweets)
```

```python
def word_cloud(tweets):
  stopwords = set(STOPWORDS)
  wordcloud = WordCloud(background_color="white", stopwords=stopwords, random_state = 2016).generate(" ".join([tw for tw in tweets]))
  plt.figure( figsize=(10,5), facecolor='k')
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.title("Twitter WordCloud")
word_cloud(tweets)
```
